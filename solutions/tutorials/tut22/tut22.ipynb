Any dictionary naturally divides into two sets: the __keys__ (what the index names are), and the __values__ (what objects the index names point to.)  We can access these sets directly with the methods of the same name - __run the cell below__ to see them work.

## T22.2 - Common use patterns for dictionaries

As I noted in lecture, dictionaries are enormously useful for __glue code__ - they enable a lot of basic operations that can help to make your code simpler and more readable.  In the parts below, we'll run through a few example use cases where use of a dictionary is key.


### Part A: Translation

One way to think of a dictionary is as a __map__ from keys to values.  Maps naturally appear in __translations__, such as between different languages; translations match equivalent things to one another.  Here's a table of a few different words in English and French:

| English | French |
|---------|--------|
| car | voiture |
| book | livre |
| coffee | café |
| dog | chien |
| thanks  | merci |

In the cell below, __make a dictionary called `english_to_french`__ with the English words as keys and the French words as values.  You should __build the dictionary with a `for` loop__ using the lists provided.

````python
english_words = ["car", "book", "coffee", "dog", "thanks"]
french_words = ["voiture", "livre", "café", "chien", "merci"]

### BEGIN SOLUTION
english_to_french = {}
for i in range(5):
    english_to_french[english_words[i]] = french_words[i]
### END SOLUTION
````

````python
print(english_to_french)
assert english_to_french["car"] == "voiture"
assert english_to_french["thanks"] == "merci"
````

Dictionaries are _one-way_ maps, but a translation is _two-way_: we could also use our map to convert French words back to English, since the pairs are unique.  Fortunately, it's not too difficult to just __invert the dictionary__, i.e. make one that points from values to keys instead.   

__Implement the function `invert_dict(d)` below__, which should return a new dictionary where the _values_ of `d` point to its _keys_.  Then run the following cell to create and test the reverse dictionary `french_to_english`.

_(Hint: start by making a new empty dictionary.  Then use a `for` loop and the `.items()` method to iterate over all key-value pairs `(k,v)` in the original dictionary, and assign them to the empty one with values as keys.)_

````python
def invert_dict(d):
    ### BEGIN SOLUTION
    inv_d = {}
    for k, v in d.items():
        inv_d[v] = k

    return inv_d
    ### END SOLUTION
````

````python
french_to_english = invert_dict(english_to_french)
print(french_to_english)

assert french_to_english["livre"] == "book"
assert french_to_english["café"] == "coffee"
````

Note: this only makes sense if the map is truly two-way, i.e. all keys and values are matched uniquely __one-to-one__!  If we try to "invert" a dictionary with multiple keys pointing to the same value, we'll lose some information.

If you truly need a bi-directional dictionary for some application in Python (`invert_dict` was simple enough, but if we have a huge data set with millions of entries it will be a massive waste of storage!), you may want to check out [the `bidict` module](https://bidict.readthedocs.io/en/main/).

### Interlude: Joining lists with `zip`

Although doing all of the above with `for` loops is good practice for understanding what is happening, if you want to take two matching lists and join them together into a single dictionary, a much better way is with the built-in keyword `zip`.  `zip` takes two matching lists (or more, but we want two for a dictionary!), and creates a _single_ list containing tuples of pairs.  Here's a quick demonstration:

````python
list_A = [1, 2, 3, 5]
list_B = ["A", "B", "C", "Z"]

list(zip(list_A, list_B))
````

Notice that I have to use `list(zip(...))` since `zip` is another _lazy object_ - we have to force it to evaluate as a list if we want to see the whole thing at once.  But if I instead type-cast it to a `dict()`, then the tuples are interpreted as key-value pairs:

````python
dict(zip(list_A, list_B))
````

In the cell below, __re-create the dict `english_to_french`__ from the lists `english_words` and `french_words` by using `zip()` and `dict()`.  (Much easier than what you did above with `for`!  In fact, if we start with two matched lists, it's easy to make both dictionaries corresponding to mapping in either direction - in other words, it's just as easy to make `french_to_english`, we don't need the `invert_dict` function.)

````python
english_to_french = dict(zip(english_words, french_words))
print(english_to_french)
````

### Part B: Descriptive Data

Another common use case for dictionaries is storing __descriptive data__.  This comes in many forms, so it's probably best to just see an example.  Here is a data cell with some information:

````python
# m, F, mu
block_list = (
    (0.5, 0.8, 0.2),
    (1.0, 3.0, 0.2),
    (2.0, 4.4, 0.2),
    (3.0, 5.6, 0.1),
    (2.0, 2.2, 0.3),
    (2.0, 1.6, 0.1),
    (1.0, 2.6, 0.1),
)
block_color_list = ("red", "orange", "yellow", "green", "blue", "white", "black")
````

You may notice that the two lists (tuples, actually) have the same length, which indicates that they are both describing some shared set of data.  The comment indicates that the data in `block_list` are `(m, F, mu)`: short for mass, applied force, and coefficient of static friction.  (In fact, this data describes a simple experimental setup: if we apply the given force to a block with the given mass and static friction, does it move or not?)

Although the comment helps, this is an ideal use case for a dictionary - dictionaries allow us to write __self-documenting code__, in which the data storage medium tells us what is in it!



In the _second cell_ below, I've created a function `check_if_block_moves` that checks if the force of static friction is overcome given the applied force or not.  This function assumes that `blocks` contains a __list of dictionaries__, with each dictionary having keys `m`, `mu`, `F` and `color` to describe a single block.

In the _first_ cell below, __create a list of dictionaries called `blocks`__ so that the cell with `check_if_block_moves()` will run without error (and give the right answer!)  You can do this by hand if you want - it will be tedious! - or use a `for` loop to build the `blocks` list of dictionaries using the data from above.  

_(Hint: Since there are two matching lists, this calls for the __iterator__ pattern.)_  

_(Optional challenge: Actually, there's a very slick way to combine matching lists and iterate over them - we can use the `zip` built-in keyword.  If you're up for a challenge, try to use `zip` to create the `blocks` list of dictionaries without ever writing a list index!)_

````python
# By hand:
blocks = [
    {"m": 0.5, "F": 0.8, "mu": 0.2, "color": "red"},
    {"m": 1.0, "F": 3.0, "mu": 0.2, "color": "orange"},
    {"m": 2.0, "F": 4.4, "mu": 0.2, "color": "yellow"},
    {"m": 3.0, "F": 5.6, "mu": 0.1, "color": "green"},
    {"m": 2.0, "F": 2.2, "mu": 0.3, "color": "blue"},
    {"m": 2.0, "F": 1.6, "mu": 0.1, "color": "white"},
    {"m": 1.0, "F": 2.6, "mu": 0.1, "color": "black"},
]

# Or the automated version from the lists above:
blocks = []
for i in range(7):
    b = {}
    b["m"], b["F"], b["mu"] = block_list[i]
    b["color"] = block_color_list[i]
    blocks.append(b)

# And the fancy way using zip:
blocks = []
for (m, F, mu), color in zip(block_list, block_color_list):
    b = {}
    b["m"], b["F"], b["mu"], b["color"] = m, F, mu, color
    blocks.append(b)

print(blocks)
````

````python
# Run this cell after you make the blocks dictionary above!


def check_if_block_moves(block):
    g = 9.8
    F_friction = block["mu"] * block["m"] * g

    if block["F"] <= F_friction:
        print(f"The {block['color']} block doesn't move.")
    else:
        print(f"The {block['color']} block moves!")


for block in blocks:
    check_if_block_moves(block)
````

You should see the following text:

```
The red block doesn't move.
The orange block moves!
The yellow block moves!
The green block moves!
The blue block doesn't move.
The white block doesn't move.
The black block moves!
```

Notice how clean and readable the code in `check_if_block_moves()` is using dictionary-stored descriptive data!  The formula $F_f = \mu m g$, for example, is clearly written out in terms of the data keys `mu` and `m`.
