## 10.2 - Monte Carlo and moments of inertia (22 points)

A sphere of radius 6 cm is made of solid resin, but with a cylindrical gap of radius $r=$ 2 cm and height $h=4$ cm centered at $\mathbf{R} = (2,2,2)$ cm from the origin of the larger sphere:

<img src="https://raw.githubusercontent.com/wlough/CU-Phys2600-Fall2025/main/lectures/img/sphere-cylinder2.png" style="max-width:100%" />

A 3D printer is probably the only way to fabricate something like this, so we'll assume the solid part of the sphere is made from plastic resin (density $\rho \sim 1.2\ \rm{g}/\rm{cm}^3$).  We would like to know the moment of inertia $I_z$ about the $z$-axis, given by the integral

$$
I_z = \int dV\ \rho(\mathbf{r}) r^2
$$
where $r$ is the cylindrical radius, i.e. the distance to the $z$-axis.  (Although you may not have seen this integral before, you should at least remember from intro physics that the moment of inertia acts like the mass for rotational motion: if we calculate $I_z$, then for spinning about the $z$-axis the _angular momentum_ $L$ will be related to the _angular velocity_ $\omega_z$ by $L = I_z \omega_z$.)



This will be a really frustrating integral to set up in any coordinate system, because of the off-center missing spherical region.  Fortunately, it's easy to do with Monte Carlo!  We'll do hit-or-miss, defining the __indicator function__

$$
f(x,y,z) = \begin{cases}
1, & \textrm{(x,y,z) inside object;} \\
0, & \textrm{(x,y,z) outside object.}
\end{cases}
$$

Then $\rho(\mathbf{r}) = \rho \cdot f(x,y,z)$, where $\rho$ is now just a constant.  Then the integral becomes

$$
I_z = \rho \int dV\ (x^2 + y^2)\ f(x,y,z) \\
= \rho V \langle (x^2 + y^2)\ f(x,y,z) \rangle_V \\
= \frac{\rho V}{N} \sum_{i=1}^N (x_i^2 + y_i^2)\ f(x_i,y_i,z_i)
$$

drawing $N$ points randomly in the three-volume $V$ containing the object.  To sum up, we can deal with this by doing hit-or-miss with a slight modification:

1. Draw points randomly in a three-dimensional cube surrounding the object.
2. Keep a running total over all points, adding $(x^2 + y^2)$ if it is _inside_ the object, and $0$ if it is _outside_.
3. Plug in to the formula above to get $I_z$, dividing by the __total__ number of points $N$ drawn.

Let's implement this in pieces.  We'll use NumPy arrays for everything, since we'd like to be able to efficiently calculate with lots of random points at once.
