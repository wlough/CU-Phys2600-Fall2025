## Project 1: `cosmic`

<img src="https://raw.githubusercontent.com/wlough/CU-Phys2600-Fall2025/main/lectures/img/cosmic-preview.png" width=400px style="float:right;" />

- Simulate the evolution of cosmic rays in the Earth's atmosphere!
- Special relativity in action - time dilation and tracking energy+momentum are key!
- Random decays of unstable particles - an ideal use case for Monte Carlo simulation!

This project has more complicated physics to understand, but as with most Monte Carlo problems, putting the pieces together into a working combined model is the easy part.

## Project 2: `gravity`

<img src="https://raw.githubusercontent.com/wlough/CU-Phys2600-Fall2025/main/lectures/img/gravity-preview.png" width=400px style="float:right;" />

- Create a working model for the orbits of planets in the solar system!
- Write a solver for the trajectory of a small mass in the solar gravity well - ODE solution in action!
- Study the paths of comets and interstellar objects for "close approaches" to the planets!

The physics in this project is relatively simple - just Newton's law of gravitation, and some equations for planetary orbits.  But using it in practice to answer numerical questions can be difficult - numerical mistakes can be bigger than the size of the planets if you're not careful!  There is also a bit more data analysis involved.

## Project 3: `kinetic`

<img src="https://raw.githubusercontent.com/wlough/CU-Phys2600-Fall2025/main/lectures/img/kinetic-preview.png" width=400px style="float:right;" />

- Simulate motion of hard spheres with collisions
- See how thermalization occurs in statistical mechanics
- Test the limits of the ideal gas law

You will implement an algorithm for elastic particle collisions that exactly determines the collision time between pairs of particles and between particles and the walls. You will use this to study changes in momentum and energy in multi-particle systems, how non-zero particle size changes the ideal gas law, and how Brownian motion emerges from random collisions.
